Empregue o `GridSearchCV` com 5 partições e `scoring='accuracy'` para encontrar o melhor modelo entre os valores de $k$ de $4$ a $11$ e empregando as métricas euclidiana e manhattan. Verifique o modelo obtidos e o `classification_report` desse modelo.

```python


# Define a variável y
y = df["Quality_Reject"]

# Cria o modelo base
knn = KNeighborsClassifier()

# Define o grid de parâmetros
param_grid = {
    "n_neighbors": list(range(4, 12)),
    "metric": ["euclidean", "manhattan"],
}

# Cria o GridSearchCV
clf = GridSearchCV(knn, param_grid, scoring="accuracy", cv=5)

# Treina o modelo
clf.fit(X_train, y_train)

# Imprime o melhor modelo
print(clf.best_estimator_)

# Imprime o relatório de classificação
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```
