```python
import matplotlib.pyplot as plt

# Contagem de programas por gênero
contagem_generos = df['listed_in'].str.split(', ').explode().value_counts().head(10)

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
contagem_generos.plot(kind='bar', color='salmon')
plt.title('Top 10 Gêneros Mais Comuns de Programas na Netflix')
plt.xlabel('Gênero')
plt.ylabel('Número de Programas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
import numpy as np

# Remover linhas com valores nulos nas colunas 'country' e 'listed_in'
df_cleaned = df.dropna(subset=['country', 'listed_in'])

# Contagem de programas por país e gênero
contingency_table = pd.crosstab(df_cleaned['country'], df_cleaned['listed_in'])

# Calcular o coeficiente de contingência (Cramér's V)
def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    min_dim = min(confusion_matrix.shape)
    return np.sqrt(chi2 / (n * (min_dim - 1)))

cramers_v_score = cramers_v(contingency_table)

print("Coeficiente de Contingência (Cramér's V):", cramers_v_score)

print('O valor do coeficiente de contingência é aproximadamente 0.288. Isso indica uma associação moderada entre o país de origem do programa e o gênero desse programa na plataforma da Netflix.')
```

```python
import matplotlib.pyplot as plt

# Extrair o ano de 'date_added'
df['year_added'] = pd.to_datetime(df['date_added']).dt.year

# Contagem de conteúdo adicionado por ano
conteudo_por_ano = df['year_added'].value_counts().sort_index()

# Criar o gráfico de linhas
plt.figure(figsize=(10, 6))
conteudo_por_ano.plot(kind='line', marker='o', color='skyblue')
plt.title('Variação da Quantidade de Conteúdo Adicionado à Plataforma por Ano')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Conteúdo Adicionado')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Exibir o gráfico
plt.show()
```
