```python
# Importar as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import seaborn as sns
from google.colab import files

# Fazer upload do arquivo CSV do seu computador para o Google Colab
uploaded = files.upload()

# Carregar o arquivo CSV


for file_name in uploaded.keys():
    df = pd.read_csv('life expectancy.csv')

# Verificar se os dados foram carregados corretamente
print("Dados carregados com sucesso!")
print(f"Número de linhas: {df.shape[0]}")
print(f"Número de colunas: {df.shape[1]}")
```

#Analise Exploratoria da Base de dados

Aqui vou verificar os dados estatisticos da minha base de dados. Como foi proposto escolhi uma base de dados do  Kaggle a base Life expectaancy.

```python

# imprimir as primeiras 5 linhas do dataframe

print("Primeiras 5 linhas do data frame")
print(df.head())

# Imprimir as últimas 5 linhas do DataFrame
print("Ultimas  5 linhas do data frame")
print(df.tail())

# Verificar informações básicas sobre os dados
print("\nInformações básicas sobre os dados:")
print(df.info())

# Verificar estatísticas descritivas
print("\nEstatísticas descritivas:")
print(df.describe())

# Lidar com valores nulos (se houver)
# Remover linhas com valores nulos
df = df.dropna()

# Verificar se há valores nulos restantes
print("\nVerificar se há valores nulos:")
print(df.isnull().sum())

# Plotar gráficos exploratórios
# Box plot
#plt.figure(figsize=(35, 12))
#sns.boxplot(data=df)
#plt.title("Box Plot")
#plt.show()


plt.figure(figsize=(35, 12))
sns.boxplot(data=df, showfliers=False, whis=1.5)
plt.title("Box Plot")
plt.show()


# Plotar o gráfico de correlação
plt.figure(figsize=(35, 12))
sns.heatmap(df.corr(), annot=True, cmap="Blues")
plt.title("Correlação entre as variáveis")
plt.show()

# Selecionar as variáveis de interesse
df1 = df[["Life Expectancy World Bank", "Prevelance of Undernourishment", "CO2", "Health Expenditure %", "Education Expenditure %"]]

# Criar um heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(df1.corr(), annot=True, cmap="Blues")
plt.title("Correlação entre variáveis de saúde e bem-estar")
plt.show()


# Selecionar as variáveis de interesse 2
df2 = df[["Life Expectancy World Bank", "Prevelance of Undernourishment"]]

# Criar um scatterplot
plt.figure(figsize=(10, 10))
plt.scatter(df2["Life Expectancy World Bank"], df2["Prevelance of Undernourishment"])
plt.title("Relação entre expectativa de vida e prevalência de desnutrição")
plt.show()

# Selecionar a variável de interesse
df3 = df[["Life Expectancy World Bank"]]

# Criar um histograma
plt.figure(figsize=(10, 10))
plt.hist(df3["Life Expectancy World Bank"])
plt.title("Distribuição da expectativa de vida")
plt.show()

```
